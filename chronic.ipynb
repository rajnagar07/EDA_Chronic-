```python
cols = ['age','blood_glucose_random','serum_creatinine','blood_pressure','class']  ## customised pairplot
p = sns.PairGrid(df[cols],hue = 'class' ,palette="coolwarm")
p.map_upper(sns.scatterplot) #upper corer plot
p.map_lower(sns.kdeplot,color = 'Blues_d') #diagonal plot
p.map_diag(sns.histplot)#lower corener plot
p.add_legend()
plt.title("pairgrid for selected columns")
plt.show()
```

```python


# Assuming you have your DataFrame 'df' loaded
# For example: df = pd.read_csv('your_data.csv')

# 1. Calculate the correlation matrix
corr = df.corr(numeric_only=True)

# 2. Create a mask for the upper triangular portion
mask = np.triu(np.ones_like(corr, dtype=bool))

# 3. Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# 4. Generate the heatmap with improvements
sns.heatmap(corr, 
            mask=mask,           # Apply the mask
            annot=True,          # Show the correlation values
            fmt='.2f',           # Format values to 2 decimal places
            cmap='coolwarm',     # Use a diverging colormap
            linewidths=0.5,      # Add lines between cells
            annot_kws={'size': 8}) # Adjust annotation font size

# 5. Add a title for clarity
plt.title('Correlation Matrix of Features', fontsize=16)

# 6. Display the plot
plt.show()
```
